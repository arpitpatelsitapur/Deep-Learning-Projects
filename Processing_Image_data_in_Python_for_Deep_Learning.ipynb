## **2. Resizing Image using pillow library**

```python
from PIL import Image
img=Image.open('/content/dog.jpg')
img_resized=img.resize((200,200))
img_resized
```

```python
img_resized.save('dog_img_resized.jpg')
```

```python
resized_img_loaded=mpimg.imread('/content/dog_img_resized.jpg')
resized_img_loaded.shape
```

## **3. RGB to grayscale conversion using OpenCV**

```python
import cv2
img=cv2.imread('/content/dog_img_resized.jpg')
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
gray_img.shape
```

```python
gray_img
```

```python
cv2.imwrite('dog_grayscale_img.jpg',gray_img)
```
